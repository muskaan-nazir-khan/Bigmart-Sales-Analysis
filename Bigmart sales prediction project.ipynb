# Label Encoding!

```python
# Label Encoding is a technique that is used to convert categorical columns into numerical ones 
# so that they can be fitted by machine learning models 
# which only take numerical data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Le = LabelEncoder()
```

```python
data['Item_Identifier'] = Le.fit_transform(data['Item_Identifier'])
```

```python
data['Item_Fat_Content'] = data['Item_Fat_Content'].replace({'LF':0, 'Low Fat':1, 'Regular':2, 'low fat':3, 'reg':4})
```

```python
data['Outlet_Identifier'] = data['Outlet_Identifier'].replace({'OUT013':0, 'OUT018':1, 'OUT019':2, 'OUT027':3,'OUT035':4,'OUT046':5,'OUT049':6})
```

```python
data['Item_Type'] = data['Item_Type'].replace({
'Fruits and Vegetables':0, 'Snack Foods':1, 'Household':2,'Frozen Foods':3, 'Dairy':4, 'Canned':5, 'Baking Goods':6, 'Health and Hygiene':7, 'Soft Drinks':8, 'Meat':9, 'Breads':10, 'Hard Drinks':11, 'Others':12, 'Starchy Foods':13, 'Breakfast':14, 'Seafood':15
})
```

```python
data['Outlet_Size'] = data['Outlet_Size'].replace({"Small": 0, "Medium": 1, "High": 2})
data['Outlet_Location_Type'] = data['Outlet_Location_Type'].replace({"Tier 1": 0, "Tier 2": 1, 'Tier 3': 2}) 
data['Outlet_Type'] = data['Outlet_Type'].replace({"Grocery Store": 0, "Supermarket Type1":1, "Supermarket Type2":2,"Supermarket Type3":3})
```

# Multiple Linear Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
#it removes the mean and scales each feature/variable to unit variance
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
final_model=LinearRegression().fit(X_train,Y_train)
```

```python
y_pred=final_model.predict(X_test)
```

```python
pd.DataFrame({'test':Y_test,'pred':y_pred})
```

```python
newdata=[[7.234,1,0.234,4,913,1,2,0]]
final_model.predict(newdata)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,y_pred)
```

```python
import xgboost as xg #supervised ,for both classification and regression but mostly for classification
```

```python
xgb = xg.XGBRegressor(n_estimators=500) #how much estimation
```

```python
xgb.fit(X,Y)
```

```python
xgb_preds = xgb.predict(test_X)
```

```python
r2_score(test_Y, xgb_preds)
```
